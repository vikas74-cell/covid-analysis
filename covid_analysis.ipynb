```python
# COVID-19 Trends Analysis Script
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('covid_19_data.csv')

# Convert 'ObservationDate' to datetime
df['ObservationDate'] = pd.to_datetime(df['ObservationDate'])

# Group by date and calculate total cases, deaths, and recoveries
daily_cases = df.groupby('ObservationDate')['Confirmed'].sum()
daily_deaths = df.groupby('ObservationDate')['Deaths'].sum()
daily_recoveries = df.groupby('ObservationDate')['Recovered'].sum()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(daily_cases, label='Confirmed Cases', color='blue')
plt.plot(daily_deaths, label='Deaths', color='red')
plt.plot(daily_recoveries, label='Recoveries', color='green')

# Add labels and title
plt.title('COVID-19 Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)

# Save the chart
plt.savefig('covid_trends.png')
plt.show()
```
